```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = sns.load_dataset('iris')
print(df.head())
print(df.info())
```

```python
# a) Gráfico de barras
petal_means = df.groupby("species")[["petal_length", "petal_width"]].mean()
petal_means.plot(kind="bar", figsize=(8,6))
plt.title("Promedio de largo y ancho del pétalo por especie")
plt.ylabel("Medida promedio (cm)")
plt.legend(["Largo del pétalo", "Ancho del pétalo"])
plt.show()
```

```python
# b) Histograma
plt.figure(figsize=(8,6))
sns.histplot(df["sepal_length"], bins=20, kde=True, color="red")
plt.title("Distribución de la longitud del sépalo")
plt.xlabel("Longitud del sépalo")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# c) Gráfico de dispersión
plt.figure(figsize=(8,6))
sns.scatterplot(x="petal_length", y="petal_width", hue="species", data=df)
plt.title("Relación entre largo y ancho del pétalo por especie")
plt.xlabel("Largo del pétalo")
plt.ylabel("Ancho del pétalo")
plt.show()
```

```python
# d) Gráfico de cajas
plt.figure(figsize=(10,6))
df_melt = df.melt(id_vars="species", value_vars=["sepal_length", "sepal_width"], var_name="Medida", value_name="Valor")
sns.boxplot(x="species", y="Valor", hue="Medida", data=df_melt)
plt.title("Distribución de medidas del sépalo por especie")
plt.xlabel("Especie")
plt.ylabel("Valor")
plt.legend(title="Medida")
plt.show()
```

```python
# e) Gráfico de líneas
plt.figure(figsize=(10,6))
sns.lineplot(x=df.index, y="petal_length", data=df, color="skyblue")
plt.title("Tendencia de la longitud del pétalo")
plt.xlabel("Índice de observación")
plt.ylabel("Longitud del pétalo")
plt.show()
```
