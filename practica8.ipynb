```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


df = sns.load_dataset('tips')

print(df.head())
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python
# a) Mapa de Calor
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="cubehelix")
plt.title("Mapa de Calor - Correlación entre variables numéricas")
plt.show()
```

```python
# b) Diagrama de Violín
plt.figure(figsize=(8,6))
sns.violinplot(x="day", y="tip", hue="day",data=df, palette="rocket", legend=False)
plt.title("Distribución de Propinas por Día de la Semana")
plt.xlabel("Día de la semana")
plt.ylabel("Propina")
plt.show()
```

```python
# c) Gráfico de Dispersión
plt.figure(figsize=(8,6))
sns.scatterplot(x="total_bill", y="tip", hue="day", data=df, palette="icefire")
plt.title("Relación entre Cuenta Total y Propina")
plt.xlabel("Total de la cuenta")
plt.ylabel("Propina")
plt.legend(title="Día")
plt.show()
```

```python
# d) Gráfico de Barras (Bar Plot)
plt.figure(figsize=(6,5))
sns.barplot(x="sex", y="tip", hue="sex", data=df, palette="coolwarm", legend=False)
plt.title("Promedio de Propinas por Género")
plt.xlabel("Género")
plt.ylabel("Propina Promedio")
plt.show()
```
